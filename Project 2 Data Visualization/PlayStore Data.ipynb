```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("playstore.csv")
```

```python
df.head(3)
```

```python
df.drop(columns=['Current Ver','Android Ver','Last Updated'],inplace=True)
```

```python
df.head(3)
```

```python
df.drop(index=10472,inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df['Reviews'] = df['Reviews'].apply(lambda x: int(x))
```

```python
df.info()
```

```python
df['Size'] = df['Size'].apply(lambda x: round(x))
```

```python
df.isnull().sum()
```

```python
df['Rating'].fillna(method='bfill',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Type'].fillna(method='bfill',inplace=True)
```

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+',''))
```

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',',''))
```

```python
df['Installs'] = df['Installs'].astype(int)
```

```python
df.info()
```

```python
df['Type'].unique()
```

```python
df['Genres'] = df['Genres'].apply(lambda x : x.split(';')[0])
```

```python
df['Genres'].unique()
```

```python
df['Content Rating'].unique()
```

```python
df['Category'] = df['Category'].apply(lambda x: x.replace('_',' '))
```

```python
df['Category'] = df['Category'].apply(lambda x: x.title())
```

```python
df['Content Rating'].value_counts()
```

```python
df['Price'] = df['Price'].apply(lambda x: float(x.replace('$','')))
```

```python
df['Price'] = df['Price'].apply(lambda x: x*75)
```

<h1>Analysis Part

```python
df.head(3)
```

<h3>1. Count the apps based o Category

```python
sns.set_theme(font_scale=1.2)
sns.set_style('whitegrid')
plt.figure(figsize=(10,5))
sns.countplot(x='Category',data=df,color='#16a085')
plt.title('No. of Apps per Category')
plt.ylabel('No. of Apps')
plt.xticks(rotation='90')
plt.show()
```

<h4>Conclusion: Family Category has maximum number of applications

<h3>2. Total Reviews per category

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Category',y='Reviews',data=df,ci=None,color='dodgerblue')
plt.xticks(rotation='90')
plt.show()
```

<h4>Conclusion: Communication and Social category has maximum average reviews

<h3>3.Ratings distrubution

```python
plt.figure(figsize=(10,6))
sns.histplot(df,x='Rating',color='#34495e')
plt.title('Distrubution of Rating')
plt.show()
```

<h4>Conclusion: Maximum apps have ratings between 3 to 5 or we say that above 3

<h3>3. Comparison of Total apps based on Type paid or free

```python
category = df['Type'].value_counts()
```

```python
types = category.index
type_values = category.values
```

```python
plt.pie(type_values,labels=types,autopct='%1.0f%%',radius=2,colors=['#00b894','#f39c12'],explode=[0.2, 0],pctdistance=0.8)
plt.title('Ratio of Paid or free apps')
plt.show()
```

<h4>Coclusion: Almost 7% apps are paid 

<h3>4.Distibution of Ratings with Reviews

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Rating',y='Reviews',data=df)
plt.title('Reviews vs Rating flow')
plt.show()
```

<h4>Conclusion: Mostly apps with having most number of reviews have high rating chances

<h3>5. Relation between Installs and Rating

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Rating',y='Installs',data=df,hue='Type')
```

<h3>6.Average ratings by Genres

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Genres',y='Reviews',data=df,ci=None,color='#BDC581')
plt.xticks(rotation='90')
plt.show()
```

<h4>Conclusion: Strategy,Communiction and Social have max reviews on average

<h3>7.Content Rating Ratios 

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Content Rating',data=df,hue='Type')
plt.title('Content count of Apps by Type')
plt.show()
```

<h4>Conclusion: Everyone content has most number of apps
