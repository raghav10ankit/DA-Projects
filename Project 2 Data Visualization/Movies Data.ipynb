```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
df['country'].unique()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['budget'].fillna(method='bfill',inplace=True)
```

```python
df['gross'].fillna(method='bfill',inplace=True)
```

```python
df['runtime'].fillna(method='bfill',inplace=True)
```

```python
df['company'].fillna(method='bfill',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['rating'].fillna(method='ffill',inplace=True)
```

```python
df.info()
```

```python
df['released'] = df['released'].astype(str).apply(lambda x: x.split('(')[0])
```

```python
df['released'] = pd.to_datetime(df['released'])
```

```python
df['released']
```

```python
df.info()
```

```python
df['genre'].unique()
```

```python
df['rating'].unique()
```

```python
df.drop(columns=['year'],inplace=True)
```

```python
df['released'].iloc[5730] = '2010-02-12'
```

```python
df['released'].iloc[5728] = '2010-10-29'
```

```python
df['Year'] = pd.DatetimeIndex(df['released']).year
```

```python
df['Year'] = df['Year'].apply(lambda x: int(x))
```

```python
df['Year'].dtypes
```

```python
year_range = []
for i in df['Year']:
#     print(i)
    if i >= 1980 and i <= 1985:
        year_range.append('1980-1985')
    if i > 1985 and i <=1990:
        year_range.append('1986-1990')
    if i > 1990 and i <=1995:
        year_range.append('1991-1995')
    if i > 1995 and i <=2000:
        year_range.append('1996-2000')
    if i > 2000 and i <=2005:
        year_range.append('2001-2005')
    if i > 2005 and i <=2010:
        year_range.append('2006-2010')
    if i > 2010 and i <=2015:
        year_range.append('2011-2015')
    if i > 2015 and i <= 2020:
        year_range.append('2016-2020')
len(year_range )       
```

```python
df['year_range'] = year_range
```

```python
df[['Year','year_range']]
```

```python
df['year_range'].unique()
```

<h2>Visualization

<h3>1.Top 10 average grossing company

```python
company = df.groupby(by='company').mean()
```

```python
top_company = company.sort_values(by='gross',ascending=False).head(10)
top_company
```

```python
sns.set_theme(font_scale=1.2)
plt.figure(figsize=(10,6))
sns.barplot(x=top_company.index,y='gross',data=top_company,color='dodgerblue')
plt.title('Top 10 Company with Gross')
plt.xticks(rotation='90')
plt.show()
```

<h3>2.Relation betwen Budget and Gross

```python
plt.figure(figsize=(10,6))
plt.scatter(df['budget'],df['gross'],color='red')
plt.title('Relation between Gross and Budget')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(df['budget'],df['gross'],line_kws={'color':'red'})
plt.title('Relation between Gross and Budget with Trend')
```

<h3>3.Top 5 countries with maximum number of movies

```python
country = df.groupby(by='country').count()
top_countries = country.sort_values(by='name',ascending=False).head(5)
top_countries
```

```python
plt.figure(figsize=(16,6))
sns.barplot(top_countries.index,top_countries['name'])
plt.title('Countries with maximum number of movies')
plt.ylabel('Count')
plt.show()
```

<h3>4. Effect of length of movie on gross of movie

```python
plt.figure(figsize=(10,6))
sns.scatterplot(df['runtime'],df['gross'])
plt.title('Effect of Movie length/runtime on Gross earnings')
plt.show()
```

<h4>Conclusion : Movie with length more than 300 minute didnt do good gross earnings and good earning movies have range of 150 to 200 minute runtime
